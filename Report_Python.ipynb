<p style="font-family: Calibri, serif; text-align: left;

Điều này cho thấy rằng bản tóm tắt có thể truyền tải thông tin một cách súc tích hơn so với toàn bộ đoạn hội thoại, điều này đúng với bản chất của một bản tóm tắt.</p>
